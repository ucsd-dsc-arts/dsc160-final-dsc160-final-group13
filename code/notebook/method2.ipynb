```python
import pandas as pd
import numpy as np
import math
from miditime.miditime import MIDITime
from mido import MidiFile, Message
```

```python
# This is the extraction of the data
time = pd.read_csv("data/time_series_data.csv")
others=pd.read_csv("data/other_information.csv")
location=pd.read_csv("data/location.csv")
```

```python
# This is necessary for the input into the function
workingWith=time.T
```

```python
# This is the remove the first few headings which are not necessary 
new_header=workingWith.iloc[0]
workingWith=workingWith[1:]
workingWith.columns=new_header
```

```python
# Get the maximum value of the cases, so that we can use linear projection 
dailyGrowthMax=workingWith.diff(axis=0).max(axis=1)
maxofUSCaseGrowth=dailyGrowthMax.max(skipna=True)
```

```python
# The original dataset is the daily cumulative case count of the counties. We want daily increase.
central=workingWith.diff(axis=0)
```

```python
# Get rid of the first row becuase the diff function makes it N/A and reindex
central=central.iloc[1:,:]
central.index = np.arange(0, central.shape[0],1)
central
```

```python
# initialization of the midi class
mymidi = MIDITime(120, 'myfile.mid', base_octave=1, octave_range = 5)
noteList = []
c_major = ['C', 'D', 'E', 'F', 'G', 'A', 'B']

for index, row in central.iterrows():
    for indexELEMENT, element in row.items():
        if element > 0:
            # Assign the scale according to a linear scale between the 1 and the maximum cases, starting from the highest (pitch) note to lowest (pitch) note
            scale = mymidi.linear_scale_pct(1, maxofUSCaseGrowth, element, reverse=True)
            # Assign the note from the scale 
            note = (mymidi.scale_to_note(scale, c_major))
            # Midi Pitch
            midi_pitch = mymidi.note_to_midi_pitch(note)
            # Assign the ptich to the notelist
            noteList.append([index, midi_pitch, 100, 1])          
#To the track 
mymidi.add_track(noteList)
#Save to the midi file
mymidi.save_midi()   
```
